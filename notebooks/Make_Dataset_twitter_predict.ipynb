```python
import pandas as pd
import re
import os
```

```python
FOLDER_PATH = '../data/raw/twitter/zemmour'
DATE = 'start_time-2022-03-18'
```

```python
files = os.listdir(FOLDER_PATH)
nb = 1
list_dfs = []
for file in files:
    if file.endswith(".csv") and DATE in file:
        list_dfs.append(pd.read_csv(FOLDER_PATH + '/' + file, index_col=0))
df_tmp = pd.concat(list_dfs, axis=0, ignore_index=True)
```

```python
pd.set_option('display.max_rows', None)
print(df_tmp)
```

```python
print(df_tmp.head())
```

```python
df_tmp['candidate'] = 'zemmour'
df_tmp.head()
```

```python
df = df_tmp[['candidate']].copy()
df['time'] = df_tmp['created_at'].map(lambda x: re.search(r'\d{2}:\d{2}:\d{2}', x).group())
df['tweet_id'] = df_tmp['id']
df['text'] = df_tmp['text']
```

```python
df.head()
```

```python
print(df)
```

```python
df = df[~df.text.str.contains("RT")]
print(df)
   
```

```python
df = df.reset_index()
print(df)
```

```python
df = df.drop('index', axis=1)
print(df)
```
